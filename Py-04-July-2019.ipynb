#### Matplotlib



```python
# Linear Graph

import matplotlib.pyplot as plt
plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()

```

```python
# Graph to plot distance and time

distance = list(range(2,11,2))
time = list(range(1,6))
# d = 2 * t
#distance
#time
plt.title("Distance VS Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,distance,"yo")
plt.grid()
plt.show()
```

```python
# v = a t
t = list(range(1,11))
a =list(range(len(t)+1))

v = [a[i] * i for i in range(len(t))]

plt.plot(t,v,'ro')
plt.grid()
plt.show()
```

```python
# Numbers VS 
n = list(range(10))
sqn = [i ** 3 for i in n]
plt.plot(n,sqn)
#plt.grid()
plt.show()
```

```python
# Voltage - Time
import numpy as np
#t = [i/10 for i in range(1,21)]
t = np.arange(0.1,2.0,0.1)
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.title("Time VS Voltage Graph")
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("Voltage")


plt.savefig("Images/Voltage-Time.png")
plt.show()
```

```python
# Plot Income of Arkanasas from 2005 to 2013 
import pandas as pd
df = pd.read_csv('Datafiles/income.csv')
years = df.columns[2:]
incomeArkanasas = df.values[3, 2:]

#incomeArkanasas

plt.title("Income of Arkanasas")
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.plot(years,incomeArkanasas)
plt.show()
```

```python
# Income of all states from 2005  to 2013
df
incomeallstates = [] 

for i in (df.columns[2:]):
    print(i)
    incomeallstates.append(df[i].mean())
print(incomeallstates)
plt.title("All States Income ")
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.plot(years,incomeallstates)
plt.show() 
```
